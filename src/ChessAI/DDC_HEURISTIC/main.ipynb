```python
from engine import SuperSeiChessEngine
from evaluation.heuristic_engine import HeuristicEngine
import chess

engine = SuperSeiChessEngine('./config/super_sei_configs.yaml')
enemy = HeuristicEngine()
board = chess.Board()

def tournament(n_games: int = 10):
    total = {'win': 0, 'draw': 0, 'loss': 0}

    for i in range(n_games):
        print(f"\n===== Game {i+1}/{n_games} =====")
        board.reset()
        last_move = None

        # Đảo màu giữa các ván để công bằng
        if i % 2 == 0:
            white_engine, black_engine = engine, enemy
        else:
            white_engine, black_engine = enemy, engine

        while not board.is_game_over():
            if board.turn == chess.WHITE:
                if white_engine == engine:
                    print(f"Last Move: {last_move}")
                    move = white_engine.get_best_move(board, last_move)
                    print(f"SuperSei (White) plays: {move.uci()}")
                else:
                    move = white_engine.get_best_move(board, depth=2)
                    print(f"Heuristic (White) plays: {move.uci()}")
            else:
                if black_engine == engine:
                    print(f"Last Move: {last_move}")
                    move = black_engine.get_best_move(board, last_move)
                    print(f"SuperSei (Black) plays: {move.uci()}")
                else:
                    move = black_engine.get_best_move(board, depth=2)
                    print(f"Heuristic (Black) plays: {move.uci()}")

            board.push(move)
            last_move = move.uci()

        # Ghi kết quả
        res = board.result()
        print(f"Final Result: {res}")

        if res == '1-0':
            if white_engine == engine:
                total['win'] += 1
            else:
                total['loss'] += 1
        elif res == '0-1':
            if black_engine == engine:
                total['win'] += 1
            else:
                total['loss'] += 1
        else:
            total['draw'] += 1

        print("Game Over.")

    print("\n===== Tournament Finished =====")
    print(f"SuperSei Results out of {n_games} games:")
    print(f"Wins: {total['win']}")
    print(f"Losses: {total['loss']}")
    print(f"Draws: {total['draw']}")

# Run the tournament
tournament(5)
```
